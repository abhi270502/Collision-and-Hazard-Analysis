```python
from ultralytics import YOLO
import torch
import os

# ---------------------------
# CONFIG
# ---------------------------
DATA_PATH = "/absolute/path/to/BDD_YOLO/data.yaml"
MODEL = "yolov8m.pt"   # use s or m (not n)
EPOCHS = 100
IMG_SIZE = 640
BATCH_SIZE = 32

# ---------------------------
# DEVICE
# ---------------------------
device = 0 if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# ---------------------------
# LOAD MODEL
# ---------------------------
model = YOLO(MODEL)

# ---------------------------
# TRAIN
# ---------------------------
results = model.train(
    data=DATA_PATH,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    device=device,

    # -------------------
    # ROBUST TRAINING
    # -------------------
    patience=30,
    cos_lr=True,
    close_mosaic=10,

    # -------------------
    # AUGMENTATION
    # -------------------
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=10,
    translate=0.1,
    scale=0.5,
    shear=2.0,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.2,

    # -------------------
    # HANDLE IMBALANCE
    # -------------------
    cls=2.0,

    # -------------------
    # LOGGING
    # -------------------
    project="runs/train",
    name="bdd_local",

    # -------------------
    # PERFORMANCE
    # -------------------
    workers=8,
    amp=True,   # mixed precision
)

print("Training completed!")
```
